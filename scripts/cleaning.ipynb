```python
import pandas as pd
import ast 
```

```python
df_raw = pd.read_csv('../data/scraped/raw_data.csv')
df_raw.head(1)
```

```python
df_raw.shape
```

```python
df = df_raw.copy()
```

```python
df.isnull().sum()
```

```python
df = df[~df.answer.isnull()]
```

```python
df.id = df.id.apply(lambda x: x.strip('ID: ')).astype(int)
```

```python
df[['name', 'question', 'answer', 'email']] = df[['name', 'question', 'answer', 'email']].map(lambda x: x.strip())
```

```python
import ast 
df[['answer_date', 
    'read_count', 
    'average_rate']] = df['meta'].apply(ast.literal_eval).apply(lambda x: (x[0], 
                                                                           x[1], 
                                                                           x[2])).apply(pd.Series)

df[['answer_date', 'read_count', 'average_rate']]
```

```python
df[['answer_date', 
    'read_count', 
    'average_rate']] = df['meta'].apply(ast.literal_eval).apply(lambda x: (pd.to_datetime(x[0], dayfirst = True), 
                                                                           int(x[1].strip('Oxunub: ')), 
                                                                           int(x[2].strip('Orta qiym…ôt: ')))
                                                                ).apply(pd.Series)

df[['answer_date', 'read_count', 'average_rate']]
```

```python
df['rate_count'] = df['rates'].apply(ast.literal_eval).apply(lambda x: x[1]).str.extract(r'(\d+)').astype(int)
```

```python
df.drop(['rates', 'meta',  'email'], axis = 1, inplace = True)
```
